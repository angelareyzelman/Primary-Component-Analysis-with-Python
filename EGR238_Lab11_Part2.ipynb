```python
import numpy as np
import matplotlib.pyplot as plt
import csv
#%% question 1
x = []
y = [] 
    
data = np.loadtxt('PCAdata_Y_good_2D.txt', delimiter=',')
x = data[:,0].copy()
y = data[:,1].copy()
    

fig = plt.subplots()
plt.scatter(x,y)
plt.xlim(xmin = -10, xmax=10)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Y_good_2D')

xm = np.mean(x)
ym = np.mean(y)

xcent = x-xm
ycent = y-ym

fig = plt.subplots()
plt.scatter(xcent,ycent)
plt.xlim(xmin = -10, xmax=10)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Y_good_2D mean-centered')

Z = np.column_stack((xcent, ycent))
theta = np.linspace(0, np.pi*2, 200)
u = np.column_stack((np.cos(theta), np.sin(theta)))
proj = 0
zi = 0
Fz = []
for i in range(len(u)):
    m = u[i]
    for i in range(len(x)):
        proj += (np.dot(m, Z[i]))**2
        zi += np.dot(Z[i], Z[i])
    Fz.append((proj)/(zi))
    proj=0
    zi=0
#print(Fz)

fig = plt.subplots()
plt.plot(np.linspace(0,2*np.pi,200),Fz, '-')
plt.xlabel('Theta (rad)')
plt.ylabel('Fz(u)')
plt.title('Fz(u) for Y_good_2D vs theta')
```
